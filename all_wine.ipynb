```python
#1 The first plot you should do is a pie chart, showing the quantity of red wines compared to the quantity of white wines.
amount_redwine = df_red["quality"].value_counts().sum()
amount_whitewine = df_white["quality"].value_counts().sum()
y = np.array([amount_redwine, amount_whitewine])
mylabels = ["red wine", "white wine"]
mycolors = ["crimson", "#17becf"]
explode = [0.05,0]

#df_red.count() # 1599
#df_white.count() # 4898

#fig = plt.figure()
plt.figure(figsize=(8, 8), facecolor = "w")

plt.rcParams['text.color'] = 'darkblue'
plt.pie(y, labels = mylabels, colors = mycolors, explode = explode, autopct='%1.0f%%', startangle = 0 )
plt.title("Quantity of red and white wine")
plt.legend()
plt.show()
```

```python
#2 create a plot to show the distribution of qualities. It should look similar to the one below

white_qualityl_count = df_white["quality_label"].value_counts()
plt.figure(figsize=(10, 6)) # Size of the plot

plt.bar(white_qualityl_count.index, white_qualityl_count, width=0.9, color="#17becf")

plt.title("Quality of white wine")
plt.xlabel("quality")
plt.ylabel("Count")
plt.xticks()
plt.show()
```

```python
# White wine correlation

plt.figure(figsize= (24,18))
mask = np.triu(np.ones_like(df_white.corr(), dtype=np.bool))
sns.heatmap(df_white.corr(), vmin=-1, vmax = 1, mask = mask, annot=True, cmap="seismic")
plt.xticks(rotation=45)
plt.title("correlation white wine")
plt.show()
```

Free and Total sulfur dioxide are correlated highly but they don’t correlate with any other variables.
Residual sugar has a strong correlation to density 
Alcohol and density has strong negative correlation

```python
# red wine correlation

plt.figure(figsize= (24,18))
mask = np.triu(np.ones_like(df_red.corr(), dtype=np.bool))
sns.heatmap(df_red.corr(), vmin=-1, vmax = 1, mask = mask, annot=True, cmap = "PiYG")
plt.xticks(rotation=45)
plt.title("correlation red wine")

plt.show()
```

Free and Total sulfur dioxide are correlated highly but they don’t correlate with any other variables.
citric acid and fixed acidity are correlated highly
ph and fixed acidity are highly correlated
