```python
# The first plot you should do is a pie chart, showing the quantity of red wines compared to the quantity of white wines.

y = np.array([1599, 4898])
mylabels = ["red wine", "white wine"]
mycolors = ["crimson", "#17becf"]

#df_red.count() # 1599
#df_white.count() # 4898

#fig = plt.figure()
plt.figure(figsize=(10, 10), facecolor = "w")

plt.rcParams['text.color'] = 'darkblue'
plt.pie(y, labels = mylabels, colors = mycolors, autopct='%1.0f%%', startangle = 90 )
plt.title("Quantity of red and white wine")
plt.legend()
plt.show()
```
