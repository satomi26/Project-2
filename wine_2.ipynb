```python
# residual sugar content in wine with histogram

figure, axs = plt.subplots(1, 2, sharey= "row", figsize =(10,7))

sugar_white = df_white["residual sugar"]
sugar_red = df_red["residual sugar"]  
# For Sine Function

axs[0].set_title("residual sugar white")
axs[1].set_title("residual sugar red")

axs[0].hist(sugar_white, fill = False, edgecolor = "black")
axs[1].hist(sugar_red, color= "red")

plt.ylim(0, 3500)
plt.suptitle("Residual sugar content in wine")

plt.xticks(np.arange(0, 15,2.5))
plt.ylabel("Frequency")
plt.show()
```

```python

# same graph with seaborn and subplot number)
plt.figure(figsize=(12, 8))
plt.subplot(121, title = "residual sugar white")
sns.histplot(df_white["residual sugar"], fill = False, edgecolor = "black")
plt.ylabel("Frequency")
plt.subplot(122, title = "residual sugar red")
sns.histplot(df_red["residual sugar"], color = "crimson")
plt.ylabel("Frequency")
#plt.xticks(np.arange(0, 15,2.5))

#plt.subplots_adjust(wspace = 0.5)
plt.suptitle("Residual sugar content in wine")
#plt.subplots_adjust(wspace = 0)
plt.show()

plt.show()
```

```python
#plt.figure(figsize=(16,16))
sns.jointplot(data = df_red, x="quality", y="sulphates", color = "crimson",  marginal_ticks=True,  height=8, ratio=4)
plt.suptitle("Red wine Sulphates - Quality")
plt.subplots_adjust(top= 0.9)
plt.show()
```
